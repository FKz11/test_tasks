```python
strin = '1 1 3 3 3 4'
inter = [int(n) for n in strin.split()]
up_sort = inter.copy()
up_sort.sort()
low_sort = inter.copy()
low_sort.sort()
flag_1 = True
for i in range(len(inter)):
  if inter[i] != up_sort[i]:
    flag_1 = False
    break
flag_2 = True
for i in range(len(inter)):
  if inter[i] != low_sort[i]:
    flag_2 = False
    break
if flag_1 or flag_2:
  print('YES')
else:
  print('NO')
```

```python
def prov(arr):
  dct = dict()
  for n in arr:
    if str(n) in list(dct.keys()):
      dct[f'{n}'] = dct[f'{n}'] + 1
    else:
      dct[f'{n}'] = 1
  if max(dct.values()) == min(dct.values()):
    return True
  else:
    return False

len_inter = ''
strin = '1 2 3 1 2 2 3 3 3 1 4 4 5'
inter = [int(n) for n in strin.split()]
flag = False
for l in range(len(inter), 2 - 1, -1):
  splc = inter[:l]
  for i in range(l):
    low_splc = splc[:i] + splc[i+1:]
    if prov(low_splc):
      print(l)
      flag = True
      break
  if flag:
      break
```

```python
def decomp(arr):
  res = []
  for l in range(1, len(arr) + 1):
    for i in range(len(arr) + 1 - l):
      res.append(arr[i:i + l])
  return res

len_inter = ''
strin = '-1 1 2 -3 6'
inter = [int(n) for n in strin.split()]
summ = 0
for sub in decomp(inter):
  for sub_sub in decomp(sub):
    if sum(sub_sub) == 0:
      summ += 1
      break
print(summ)
```

```python
n, max_sum = (int(num) for num in input().split())
ls_data = []
for i in range(n):
  ls_data.append([int(num) for num in input().split()] + [0])
ls_data.sort(key = lambda x: x[1])
i_med = n // 2
sum_half_max = max_sum - sum(map(lambda x: x[0], ls_data[:i_med]))
ls_data_half_max = ls_data[i_med:]
start_val = ls_data_half_max[0][1]
for ls in ls_data_half_max:
  if ls[0] > start_val:
    ls[2] = ls[0]
  else:
    ls[2] = start_val
while sum(map(lambda x: x[2], ls_data_half_max)) > sum_half_max:
  for ls in ls_data_half_max:
    if ls[2] > ls[0]:
      ls[2] = ls[2] - 1
      break
  ls_data_half_max.sort(key = lambda x: x[2])
  for ls in ls_data_half_max:
    if ls[0] > ls_data_half_max[0][2]:
      ls[2] = ls[0]
  else:
    ls[2] = ls_data_half_max[0][2]
print(ls_data_half_max[0][2])
```
